```python
def Power(exponent):
    print(2**exponent)
Power(3)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Power2(base,exponent):
    print(base**exponent)

Power2(3,3)
```

```python
def Power3(base,exponent):
    return base**exponent
x = Power3(3,3)
print(x)
```

```python
x, y = [] , []
for i in range(1,11):
    x.append(i)
    y.append(Power3(i,2))

plt.plot(x,y)
plt.xlabel(xlabel='x')
plt.ylabel(ylabel='y')
plt.yscale('log')
plt.show()
```

```python
def PlotPower(bases,exponent):
    x,y = [], []
    for base in bases:
        x.append(base)
        y.append(base**exponent)
    plt.plot(x,y)
    plt.xlabel(xlabel = 'x')
    plt.ylabel(ylabel='y')
    plt.yscale('log')
    plt.show()
PlotPower(np.arange(1,13),2)
```
